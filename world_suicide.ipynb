### **ğŸ˜ Coding**

```python
# Data Loading
import pandas as pd
df = pd.read_excel(r"data\world_suicide.xlsx")
df.head()

# 1. Check for Duplicate Entries
duplicates = df.duplicated().sum()
print("No of duplicate rows: ", duplicates)
df[df.duplicated()].head()

# 2. Check for missing values
missing_col = df.isnull().sum()
print("Missing values per column:\n", missing_col)
total_miss = missing_col.sum()
print("\nTotal missing values: ", total_miss)


# 3. Nigeria's highest suicide motality rate
# filter for nigeria rows:
nigeria_df = df[df["Country Name"] == "Nigeria"]
nigeria_df
if not nigeria_df.empty:
    nigeria_highest_row = nigeria_df.loc[nigeria_df["Suicide Rate"].idxmax()]
    print(f"\nNigeria's highest suicide rate:\n {nigeria_highest_row}")
else:
    print("Nigeria not found in Dataset.")

# 4. No of countries in the dataset
num_ctry = df["Country Name"].nunique()
print(f"\nNumber of unique countries: {num_ctry}")

# 5. Country with the highest SMR
overall_highest_row = df.loc[df["Suicide Rate"].idxmax()]
print(f"\n Country with the highest Suicide Rate:\n{overall_highest_row}")

# 6. Compare NIGERIA VS NIGER(Year 2015)
nigeria_2015 = df[(df["Country Name"] == "Nigeria") & (df["Year"] == "2015")] ["Suicide Rate"]
niger_2015 = df[(df["Country Name"] == "Niger") & (df["Year"] == "2015")] ["Suicide Rate"]

print("Nigeria 2015:", float(nigeria_2015.iloc[0]) if not nigeria_2015.empty else "No data")
print("Niger 2015:", float(niger_2015.iloc[0]) if not niger_2015.empty else "No data")

# 7. Highest Suicide Rate in COVID-19 year(2020)
covid_df = df[df["Year"] == 2020]
covid_highest_row = covid_df.loc[covid_df["Suicide Rate"].idxmax()]
print(f"\nHighest suicide rate in 2020 (COVID-19):\n {covid_highest_row}")

# 8. Highest suicide rate for each country
highest_per_ctry = (df.groupby("Country Name")["Suicide Rate"].max().reset_index()
                    .rename(columns={"Suicide Rate":"Max Suicide Rate"}))
display(highest_per_ctry.head(10))

# 9. Countries with Suicide Rate > 20 btw 2005 & 2010
filtering = df[
    (df["Year"] >= 2005) & (df["Year"] <= 2010) & (df["Suicide Rate"] > 20)
]
ctry_get20 = filtering["Country Name"].nunique()
print("No of Countries with rate > 20 btw 2005 & 2010: ", ctry_get20)

#10. ADDITIONAL QUESTION
# What is the trend of Nigeria Suicide Rate?
nigeria_trend = nigeria_df[
    ["Year","Suicide Rate"]
].sort_values("Year")
nigeria_trend
```
